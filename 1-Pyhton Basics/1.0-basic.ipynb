## Indentation
Indentation in Python is a fundamental aspect of the language that defines code blocks, replacing the need for braces {} used in languages like C or Java.  It is not just for readability—it is syntactically required.  Python uses consistent whitespace (spaces or tabs) at the beginning of lines to group statements into blocks, such as those within if, for, while, def, or class statements.


```python
## Indentation
## Python uses indentation to define blocks of code. The standard indentation is 4 spaces. or a tap is required

age=32
if age>30:
    
    print(age)

print(age)
```
