# Variables

Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In Python,variables are created when you assign a value to them,and they do not need explicit declaration to reserve memory space.The declaration happens automatically when you assign a value to a variable.

```python
## Declaring and Assigning Variables
age = 32
height = 6.1
name = "John"
is_student = True

## printing the variables
print("age:",age)
print("height:",height)
print("name:",name)
print("is_student:",is_student)
```

```python
## Naming Conventions
## Valid variable names should be descriptive and underscores
## variables name case sensitive
first_name = "Aasim"
last_name = "Mansuri"
print("First Name:",first_name)
print("Last Name:",last_name)
```

```python
## Understanding Variable Types
## Python is dynamically typed,type of a variable is determined at runtime
x = 10 #int
height = 5.9 #float
name = "Asim" #str 
is_student = True #bool
print(type(is_student))  # Output: <class 'bool'>
```

```python
# Type Chacking and Conversion
type(height)
```

```python
age = 25
print(type(age))  # Output: <class 'int'>

#Type conversion
age_str = str(age)
print(age_str)  # Output: '25'
print(type(age_str))  # Output: <class 'str'>
```

```python
## Dynamic Typing
## Python allow the type of a variable to change during execution
var = 10
print(var, type(var))  # Output: 10 <class 'int'>

var = "Now I'm a string"
print(var, type(var))  # Output: Now I'm a string <class 'str'

var = 3.14
print(var, type(var))  # Output: 3.14 <class 'float'>
```
